```python
import pandas as pd
import numpy as np
import wikipedia as wp
from geopy.geocoders import Nominatim
import folium

html = wp.page("List of postal codes of Canada: M").html().encode("UTF-8")

df = pd.read_html(html)[0]

##To pick index number which has borough column consists of Not assigned

nass = df[df['Borough'] == 'Not assigned'].index

##To Drop the unwanted indexs using pandas drop method

df.drop(nass, inplace=True)

print (df.head())

df.shape

```

```python
geo = pd.read_csv('Geospatial_Coordinates.csv')

merge = pd.merge(left=df,right=geo, left_on='Postcode', right_on='Postcode')

merge.head(10)
```

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
map_newyork = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(merge['Latitude'], merge['Longitude'], merge['Borough'], merge['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186fc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_newyork)  
    
map_newyork
```
